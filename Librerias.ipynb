#**2.**  Codigo generado por ChatGPT

```python
# Importar librerias
import yfinance as yf
import pandas as pd
from pandas_datareader import data as pdr
```

```python
# Sobrescribir pandas_datareader con yfinance
yf.pdr_override()
```

```python
# Lista de tickers de empresas de energias renobables
tickers = ['ENPH', 'RUN', 'FSLR', 'VWS.CO']
```

```python
# Obtener datos
data = pdr.get_data_yahoo(tickers, start="2018-10-01", end="2023-10-01", interval="1mo")
```

```python
# Filtrar solo los precios de cierre ajustados y eliminar las filas vacias
adj_close = data['Adj Close'].dropna()
```

```python
# Reindexar las fechas en el formato 'Mes Año'
adj_close.index = adj_close.index.strftime('%B %Y')
```

```python
# Exportar a Excel
adj_close.to_excel("Precios_Cierre_Ajustado_Energias_Renobables.xlsx")
print("Exportado con éxito.")
```
