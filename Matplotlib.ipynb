Matplotlib es una bibliteca para la generacion de graficos a partir de datos contenidos en listas o arrays

```python
import matplotlib.pyplot as plt
```

```python
# Simbolo % ayuda a guargar graficos
%matplotlib inline
```

```python
import numpy as np
```

```python
# Linespace (inicio, fin, cuantos numeros se ¿necesitan ?)
x = np.linspace(0, 100, 40)
print(x)
```

```python
# Tomo el numero, lo elevo al cuadrado y le sumo 5
y = np.sqrt(x) + 5
print(y)
```

```python
# Genera el grafico
plt.plot(x, y)
# Colocar nombres de la grafica
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico")
# Imprime sin codigo
plt.show()
```

# Personalizacion del grafico

```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

num = np.linspace(0, 10, 100)
seno = np.sin(num)

print(seno)
font1 = {'family': 'serif', 'color':'#FF3399', 'size':15}

plt.plot(num, seno, color = '#FF3399', linewidth=5, alpha=0.3)
plt.title('Funcion Seno', fontdict= font1, loc='left')
plt.xlabel('Numeracion')
plt.ylabel('Seno')
plt.show()
```

# Matriz de ejes

```python
'''FORMA 1      subplot(columnas, filas, posicion)

plt.subplot(1, 2, 1)
plt.plot(x, y, color = 'green')
plt.title("Raiz cuadrada", color = 'green')

plt.subplot(1, 2, 2)
plt.plot(num, seno, color = 'blue')
plt.title('Funcion Seno', color = 'blue', loc='left')
plt.show()'''

'''FORMA 2'''
# EJES es un array de graficas, se puede iterar para colocar los valores o poner por separado
# subplots(columnas, filas, figsize=[ancho, alto], dpi=resolucion)
graficos, ejes =plt.subplots(nrows=2, ncols=3, figsize=[5,3], dpi=50)

ejes[0,0].plot(x, y, color = 'green')
ejes[0,0].set_title("Raiz cuadrada", color = 'green')

ejes[1,1].plot(num, seno, color = 'blue')
ejes[1,1].set_title('Funcion Seno', color = 'blue')

ejes[0,2].plot(seno,num, color = 'red')
ejes[0,2].set_title("Cuadrado de x", color = 'red')

# Funcion para espaciar las graficas
plt.tight_layout()
```

# Ejes Superpuestos con POO



> Grafica dentro de otra grafica


```python
# Crea imagen vacia
# figure(Tamaño de la imagen)
# figsize=[ancho, alto], dpi=resolucion
grafico = plt.figure(figsize=[5,3], dpi=50)

'''    [punto de inicio, punto de final]
      [xi, yi  ...  xf, yf]   '''

mayor = grafico.add_axes([0, 0, 1, 1]) # Todo se genera apartir del objeto/figura creada

mayor.plot(x, y, color = 'green')
mayor.set_title("Gráfico", color = 'green')
mayor.set_xlabel("Eje X")
mayor.set_ylabel("Eje Y")

menor = grafico. add_axes([0.5, 0.2, 0.4, 0.4]) # Prorcentajes con respecto a la grafica mayor

menor.plot(num, seno, color = 'red')
menor.set_title('Funcion Seno', color = 'red', size=10)
menor.set_xlabel('Numeracion', size=9)
menor.set_ylabel('Seno', size=9)

plt.show()
```

# Personalizacion de varias graficas dentro del mismo eje

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```



> PERSONALIZACION DE LAS LINEAS


**color:**  RGB color codes chart\
**lw:** line width > Grosor de linea\
**alpha:** Transparencia\
**ls:** linestyle > Estilo de linea
*   'solid' **-**
*   'dashed' **--**
*   'dashdot' **-,**
*   'dotted' **:**
*   'None' **blank**

> PERSONALIZACION DE LOS MARCADORES


**mfc:**  markerfacecolor > color de relleno\
**mec:** markeredgecolor > color de borde\
**ms:** markersize > tamaño de la marca\
**mew:** markersize > tamaño del borde de la marca\
**marker:** Estilo de marca\
1. **o**
2. **x**
3. **1**

```python
x = np.linspace(0, 10, 10)

grafico = plt.figure(figsize=[4,4], dpi=80) # facecolor='blue', edgecolor = 'blue'
ejes = grafico.add_axes([0, 0, 1, 1])

ejes.plot(x, 2*x, label = '2x', color = '#FF3399', lw=4, alpha=0.5, ls='dashed')
ejes.plot(x, 2+x, label = '2+x', color = 'orange', lw=3, alpha=0.8, ls='dashdot')
ejes.plot(x, x**2, label='x^2', color = '#66B2FF', marker='o', ms=8, mfc='pink',  mew=2, mec='black')
ejes.legend(loc=4)

# Limites en x & y para mejor visualizacion de la grafica
ejes.set_xlim([0,10])
ejes.set_ylim([0,25])

# rejilla(axis = {'both', 'x', 'y'})
plt.grid(axis='both')
plt.show()
```
